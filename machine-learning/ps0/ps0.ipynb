```python
import numpy as np

x = np.random.multivariate_normal(mean=[0,0], cov=[[1,0], [0,1]], size=1000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x[:,0], x[:,1])
plt.savefig('og.png')
```

```python
xdiffmean = np.random.multivariate_normal(mean=[1,1], cov=[[1,0], [0,1]], size=1000)

plt.scatter(xdiffmean[:,0], xdiffmean[:,1])
plt.savefig('xdiffmean.png')
```

```python
xdiffvar1 = np.random.multivariate_normal(mean=[0,0], cov=[[2,0], [0,2]], size=1000)

plt.scatter(xdiffvar1[:,0], xdiffvar1[:,1])
plt.savefig('xdiffvar1.png')
```

```python
xdiffvar2 = np.random.multivariate_normal(mean=[0,0], cov=[[1,0.5], [0.5,1]], size=1000)

plt.scatter(xdiffvar2[:,0], xdiffvar2[:,1])
plt.savefig('xdiffvar2.png')
```

```python
xdiffvar3 = np.random.multivariate_normal(mean=[0,0], cov=[[1,-0.5], [-0.5,1]], size=1000)

plt.scatter(xdiffvar3[:,0], xdiffvar3[:,1])
plt.savefig('xdiffvar3.png')
```
